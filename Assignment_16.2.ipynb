## Problem Statement
### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and color each data point with specific class.

```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # visuaalization
from sklearn import decomposition # for PCA
from sklearn import datasets
```

```python
#Loading iris dataset from sklearn
iris = datasets.load_iris()
data = iris.data # data
iris.feature_names #feature's name
```

```python
# Target variable
targets = iris.target
```

```python
#Principal component analysis
from sklearn.decomposition import PCA
```

```python
# Instantiate PCA with 3 principal components
pca= PCA(n_components=3)
```

```python
data.head()
```

<b>Reduce the dimenionality of the data from 4 to 3</b>

```python
# fit transform data
new_data=pca.fit_transform(data)
```

```python
new_data.shape
```

<b>Converting the transformed data to Dataframe</b>

```python
import pandas as pd
new_data = pd.DataFrame(new_data)
```

```python
new_data.head()
```

```python
new_data.columns = ["PC1", "PC2", "PC3"]
```

<b>Plotting the data in 3D</b>

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib

# we can use below code for adopting ggplot library from r to plot the styles in maplotlib
#plt.style.use('ggplot')

#for black background
plt.style.use(['dark_background'])

# for figure size
fig = plt.figure(figsize=(12,6))

ax = Axes3D(fig) # to create 3d  axis for figure

# for size of fonts of the title of the plot 
matplotlib.rcParams.update({'font.size': 25})


#3d scatter plot color filtered with targets
ax.scatter(new_data['PC1'], new_data['PC2'], new_data['PC3'],c = targets,s=20,cmap = 'inferno')


# setting lables
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

ax.set_title("3D scatter plot for IRIS data")

#adding legends
ax.legend(["PC1","PC2","PC3"])


plt.show()
```
